```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scipy.stats import ttest_ind_from_stats
```

```python
women_Shoes_Data = pd.read_csv('7003_1.csv', usecols=['brand','prices.amountMax'])
men_Shoes_Data = pd.read_csv('7004_1.csv', usecols=['brand','prices.amountMax'])

women_Shoes_Data['brand'] = women_Shoes_Data['brand'].apply(str)
women_Shoes_Data = women_Shoes_Data.rename(columns={"prices.amountMax":"Prices"})
women_Shoes_Data = women_Shoes_Data.dropna(subset = ['Prices','brand'])

men_Shoes_Data['brand'] = men_Shoes_Data['brand'].apply(str)
men_Shoes_Data = men_Shoes_Data.rename(columns={"prices.amountMax":"Prices"})
men_Shoes_Data = men_Shoes_Data.dropna(subset = ['Prices','brand'])

comb_List = [women_Shoes_Data, men_Shoes_Data]

combined_Men_Women_Shoe_Brands = pd.concat(comb_List)



# print(combined_Men_Women_Shoe_Brands)

nike_List = []
adidas_List = []

puma_List = []
boot_Barn_Holding_List = []
caleres_Lists = []
puma_Brands = ['Puma','puma','PUMA']
caleres_Brands = ['Vionic','Vince','Edelman','Ryka','Naturalizer','Famous','Scholl','Blowfish']
boot_Barn_Holding_Brands = ['Ugg','UGG','Roper','Stetson','Nocona', 'Minnetonka','Lucchese','Bryant','Twisted X','Lamo'
                            ,'Keen','Justin Boots','Frye','Free People','Ferrini','Durango','Dingo','Dan Post','Corral'
                            ,'Coconuts','Boulet','Bed Stu','Ariat']
for x in combined_Men_Women_Shoe_Brands.itertuples(index = False):
    if x[0].find('Nike') != -1 or x[0].find('nike') != -1:
        nike_List.append(x)
    elif x[0].find('didas') != -1 or x[0].find('eebok') != -1:
        adidas_List.append(x)
    elif any(q in x[0] for q in puma_Brands):
        puma_List.append(x)
    elif any(z in x[0] for z in caleres_Brands):
        caleres_Lists.append(x)
    elif any(y in x[0] for y in boot_Barn_Holding_Brands):
        boot_Barn_Holding_List.append(x)
        
nike_df = pd.DataFrame(nike_List)
adidas_df = pd.DataFrame(adidas_List)
puma_df = pd.DataFrame(puma_List)
caleres_df = pd.DataFrame(caleres_Lists)
boot_Barn_Holding_df = pd.DataFrame(boot_Barn_Holding_List)

print(adidas_df)
```

# Hypothesis Testing For the various brands 

```python
#generating the t_statistics and P_values 
df_List = [nike_df, adidas_df, puma_df,caleres_df, boot_Barn_Holding_df]
brands_list = ["Nike","Adidas","Puma","Caleres","Boots and Barn Holding"]
df_List2 = df_List

for dataframe in range(len(df_List)):
    mean_1 = df_List[dataframe].mean()[0]
    std_1 = df_List[dataframe].std()[0]
    nobs_1 = df_List[dataframe].count()[0]
    for df in range(dataframe+1, len(df_List)):
        if dataframe == 4:
            break
        else:
            mean_2 = df_List[df].mean()[0]
            std_2 = df_List[df].std()[0]
            nobs_2 = df_List[df].count()[0]
            t_statistic,p_value= ttest_ind_from_stats(mean1=mean_1, std1=mean_1, nobs1=nobs_1, 
                                          mean2=mean_2, std2=std_2, nobs2=nobs_2,equal_var = False)
            print("The T-statistic for ",brands_list[dataframe]," and ", brands_list[df], " is: ", t_statistic)
            print("The P-Value for ",brands_list[dataframe]," and ", brands_list[df], " is: ", p_value ,"\n")
            
        
```
