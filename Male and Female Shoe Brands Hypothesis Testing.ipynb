```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Male

```python
male_Shoes_Data = pd.read_csv('7004_1.csv', usecols=['brand','prices.amountMax'])
male_Shoes_Data['brand'] = male_Shoes_Data['brand'].apply(str)
male_Shoes_Data = male_Shoes_Data.rename(columns={"prices.amountMax":"Prices"})
male_Shoes_Data = male_Shoes_Data.dropna(subset = ['Prices','brand'])
# print(male_Shoes_Data)
nike_List = []
boot_Barn_Holding_List = []
caleres_Lists = []
caleres_Brands = ['Vionic','Vince','Edelman','Ryka','Naturalizer','Famous','Scholl','Blowfish']
boot_Barn_Holding_Brands = ['Ugg','UGG','Roper','Stetson','Nocona', 'Minnetonka','Lucchese','Bryant','Twisted X','Lamo'
                            ,'Keen','Justin Boots','Frye','Free People','Ferrini','Durango','Dingo','Dan Post','Corral'
                            ,'Coconuts','Boulet','Bed Stu','Ariat']
for x in male_Shoes_Data.itertuples(index = False):
    if x[0].find('Nike') != -1 or x[0].find('nike') != -1 or x[0].find('NIKE') != -1:
        nike_List.append(x)
    elif any(z in x[0] for z in caleres_Brands):
        caleres_Lists.append(x)
    elif any(y in x[0] for y in boot_Barn_Holding_Brands):
        boot_Barn_Holding_List.append(x)
        
nike_df = pd.DataFrame(nike_List)
caleres_df = pd.DataFrame(caleres_Lists)
boot_Barn_Holding_df = pd.DataFrame(boot_Barn_Holding_List)
```

# WOMEN

```python
women_Shoes_Data = pd.read_csv('7003_1.csv', usecols=['brand','prices.amountMax'])
women_Shoes_Data['brand'] = women_Shoes_Data['brand'].apply(str)
women_Shoes_Data = women_Shoes_Data.rename(columns={"prices.amountMax":"Prices",})
women_Shoes_Data = women_Shoes_Data.dropna(subset = ['Prices','brand'])
# print(women_Shoes_Data)
nike_List = []
adidas_List = []
puma_List = []
boot_Barn_Holding_List = []
caleres_Lists = []
caleres_Brands = ['Vionic','Vince','Edelman','Ryka','Naturalizer','Famous','Scholl','Blowfish']
boot_Barn_Holding_Brands = ['Ugg','UGG','Roper','Stetson','Nocona', 'Minnetonka','Lucchese','Bryant','Twisted X','Lamo'
                            ,'Keen','Justin Boots','Frye','Free People','Ferrini','Durango','Dingo','Dan Post','Corral'
                            ,'Coconuts','Boulet','Bed Stu','Ariat']
for x in women_Shoes_Data.itertuples(index = False):
    if x[0].find('Nike') != -1 or x[0].find('nike') != -1:
        nike_List.append(x)
    elif x[0].find('didas') != -1 or x[0].find('eebok') != -1:
        adidas_List.append(x)
    elif x[0].find('Puma') != -1 or x[0].find('puma') != -1:
        puma_List.append(x)
    elif any(z in x[0] for z in caleres_Brands):
        caleres_Lists.append(x)
    elif any(y in x[0] for y in boot_Barn_Holding_Brands):
        boot_Barn_Holding_List.append(x)
        
nike_df1 = pd.DataFrame(nike_List)
adidas_df = pd.DataFrame(adidas_List)
puma_df = pd.DataFrame(puma_List)
caleres_df1 = pd.DataFrame(caleres_Lists)
boot_Barn_Holding_df1 = pd.DataFrame(boot_Barn_Holding_List)
```

```python
df_female = pd.concat([nike_df1,adidas_df,puma_df,caleres_df1,boot_Barn_Holding_df1])
df_female['Prices'] = df_female['Prices'].astype('float')
df_female.describe()
```

```python
avg_mf = {"Male": 0, "Female": 0}
avg_mf["Male"] = df['Prices'].mean()
avg_mf["Female"] = df_female['Prices'].mean()
```

```python
avgs = pd.DataFrame.from_dict(avg_mf, orient='index', columns = ["Avg Prices"])
plt.figure(figsize=(11, 7), dpi=60)
plt.bar(avgs.index.to_list(), avgs['Avg Prices'], color='blue')
plt.xticks(rotation = 'vertical')
plt.title('Average prices for Male and Female', fontsize=14)
plt.ylabel('Average Prices', fontsize=14)
plt.show()
```
